```python
# English to Marathi Model

from transformers import MarianMTModel, MarianTokenizer

model_name_mr = "Helsinki-NLP/opus-mt-en-mr"

tokenizer_mr = MarianTokenizer.from_pretrained(model_name_mr)
model_mr = MarianMTModel.from_pretrained(model_name_mr)

def translate_mr(text):
    inputs = tokenizer_mr(text, return_tensors="pt", padding=True)
    translated = model_mr.generate(**inputs)
    return tokenizer_mr.decode(translated[0], skip_special_tokens=True)

print("Marathi model loaded successfully ✅")
```

```python
text = "Artificial Intelligence improves agriculture productivity."

print("Marathi Translation:")
print(translate_mr(text))
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "facebook/nllb-200-distilled-600M"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def translate(text, target_lang):
    inputs = tokenizer(text, return_tensors="pt")
    translated_tokens = model.generate(
        **inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id[target_lang]
    )
    return tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
```

```python
text = """
Artificial Intelligence is transforming agriculture by improving productivity,
predicting weather conditions, detecting plant diseases, and automating irrigation systems.
AI helps farmers make better decisions using data analysis and smart technologies.
It reduces crop losses and increases efficiency. With AI-powered tools and drones,
farmers can monitor crops in real time and optimize resource usage.
"""

# Marathi code: mar_Deva
marathi = translate(text, "mar_Deva")

# Hindi code: hin_Deva
hindi = translate(text, "hin_Deva")

print("Marathi Translation:")
print(marathi)

print("\nHindi Translation:")
print(hindi)
```

```python
from transformers import pipeline

# Load summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

print("Summarization model loaded successfully ✅")
```

```python
text = """
Artificial Intelligence is transforming agriculture by improving productivity,
predicting weather conditions, detecting plant diseases, and automating irrigation systems.
AI helps farmers make better decisions using data analysis and smart technologies.
It reduces crop losses and increases efficiency. With AI-powered tools and drones,
farmers can monitor crops in real time and optimize resource usage.
"""
```

```python
summary = summarizer(text, max_length=60, min_length=20, do_sample=False)

print("Generated Summary:")
print(summary[0]['summary_text'])
```
